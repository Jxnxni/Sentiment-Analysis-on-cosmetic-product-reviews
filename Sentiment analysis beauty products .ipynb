```python
#open the excel file in jupyter notebook 
import pandas as pd

df = pd.read_csv(r'/Users/jananinadarajan/Documents/Personal projects /Cosmetic/nyka_top_brands_cosmetics_product_reviews.csv')
print(df)

#convert to lower
df['review'] = df['review_text'].astype(str).str.lower()
```

```python
import numpy as np

def create_sentiment(rating):
    
    if float(rating) >= 1 and float(rating) <= 2:
        return -1 # negative sentiment
    elif float(rating) >= 4 and float(rating) <= 5:
        return 1 # positive sentiment
    else:
        return 0 # neutral sentiment

df['Sentiment'] = df['review_rating'].apply(create_sentiment)

df.head()
```

```python
#tokenizaion
import nltk 

from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

df['review_token']=df['review'].apply(regexp.tokenize)
df.head(50)
```

```python
#remove stopwords #include request and access and able apply to all 
nltk.download('stopwords')

from nltk.corpus import stopwords
# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")


# Remove stopwords
df['review_token'] = df['review_token'].apply(lambda x: [item for item in x if item not in stopwords])


import re 

# Define a function to remove numbers, special characters from a tokenized list
def keep_words(tokens):
    return [token for token in tokens if re.match(r'^[a-zA-Z]+$', token)]

# Apply the function to the tokenized list column of your DataFrame
df['review_token'] = df['review_token'].apply(keep_words)


df_lem = df
#lemmatization/stemmming of words 
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    words = [lemmatizer.lemmatize(word, pos='v') for word in text]
    sentence = ' '.join(words)
    return sentence


df_lem['review_lem'] = df_lem['review_token'].apply(lemmatize_words)

df_lem['review_lem'] = df['review_lem'].str.split(' ')
# #Remove numbers

df_lem.head()
```

```python
#plot frequency
#calculate freq_dist of keywords
from nltk.probability import FreqDist #export as csv and remove stuff similar to word cloud

#df_lem['Reason_custom_token_str'] = df['Reason_custom_token'].apply(lambda x: ' '.join(x))

df_lem['review_lem_str'] = df_lem['review_lem'].apply(lambda x: ' '.join(x))

#df_lem.head()
#combine all words after removing custom stopwords into a string: every_word
every_word = " ".join([word for word in df_lem['review_lem_str']])

every_word_list = every_word.split(" ")

num_words = FreqDist(every_word_list)


# #frequency of keywords 
df_freq = pd.DataFrame({'Word': num_words.keys(), 'Frequency':num_words.values()})
df_freq = df_freq.sort_values(['Frequency'], ascending=False)

freq = df_freq[df_freq['Frequency'] >1]

freq_top_10 = df_freq[df_freq['Frequency'] >1].head(25)

#df_freq.to_csv(r'/Users/jananinadarajan/Documents/SMU/HEM_internship/NLP/freq_5_0327.csv')
from matplotlib import pyplot as plt
plt.figure(figsize=(11,7))
plt.bar(freq_top_10['Word'], 
        freq_top_10['Frequency'])

plt.xticks(rotation = 45)

plt.xlabel('Most common Words')
plt.ylabel("Frequency")
plt.title("Frequency distribution of the 25 most common words")
   
plt.show()
```

```python
#Inverse document frequency to remove naturally occuring words like "a", "the" etc.
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(strip_accents=None, 
                        lowercase=False,
                        preprocessor=None)


#transformed into a vector 
X = tfidf.fit_transform(df_lem['review_lem_str'])
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
y = df_lem['Sentiment'] # target variable
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#Fit a logistic regression classfifier on training dataset
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear')
lr.fit(X_train,y_train) # fit the model
preds = lr.predict(X_test) # make predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(preds,y_test)
```
